```python
import matplotlib.pyplot as plt
import numpy as np

# n/totient(n) function

def ntotient(n):
    resultu = 1.0  
    resultd = 1.0 

    def prime_factors(n):
        # Function to return the prime factors of n
        factors = []
        # Check for number of 2s that divide n
        while n % 2 == 0:
            if 2 not in factors:
                factors.append(2)
            n //= 2
        # n must be odd at this point, so a skip of 2 (i.e., check only odd numbers) is sufficient
        for i in range(3, int(n**0.5) + 1, 2):
            while n % i == 0:
                if i not in factors:
                    factors.append(i)
                n //= i
        # This condition is to check if n is a prime number greater than 2
        if n > 2:
            factors.append(n)
        return factors

    prime_factors_list = prime_factors(n)
    for i in prime_factors_list:
        resultu *= i
        resultd *= (i - 1)
    result = resultu / resultd

    return result
```

```python
print( ntotient(12) )
```

```python
n= 10**3

# Calculate the value of x/totient(x)
x = np.arange(2, n + 1)
y_values = np.array([ntotient(x) for x in x])
y = y_values
```

```python
max = 0
for i in range(len(y)):
    if y[i] >= max:
        max = y[i]
        index = i
        print(f"Max: {max} at x = {x[i]}")

plt.figure(figsize=(10, 5))

plt.plot(x, y)
plt.grid()
plt.show()
```

```python
min = float('inf')
for i in range(len(y)):
    if y[i] <= min:
        min = y[i]
        index = i
        print(f"Min: {min} at x = {x[i]}")

plt.figure(figsize=(10, 5))

plt.plot(x, y)
plt.grid()
plt.show()
```
